```python
df.info()
```

```python
##vehicle image delete 
df.drop(columns=["Vehicle Images"], inplace=True)
```

```python
# Customer cancellation reason
df["Canceled_Rides_by_Customer"] = df["Canceled_Rides_by_Customer"].fillna("Not Cancelled")

# Driver cancellation reason
df["Canceled_Rides_by_Driver"] = df["Canceled_Rides_by_Driver"].fillna("Not Cancelled")

# Incomplete ride reason
df["Incomplete_Rides_Reason"] = df["Incomplete_Rides_Reason"].fillna("Not Applicable")

# Payment method
df["Payment_Method"] = df["Payment_Method"].fillna("Not Applicable")
```

```python
##veryfy null
df.isnull().sum()
```

(Summary)

The missing values were analyzed based on business rules instead of blindly filling them. Cancellation reasons and payment details were populated with meaningful labels ("Not Cancelled", "Not Applicable") where appropriate. Operational metrics such as TAT and ratings were left as NULL because they are not generated for cancelled or incomplete rides. The Vehicle Images column was removed since it contained no data across all records.

Feature Engineering

```python
##Booking Month
df["booking_month"] = df["Date"].dt.month_name()
```

```python
##Booking Day
df["booking_day"] = df["Date"].dt.day_name()
```

```python
##Booking Hour
df["booking_hour"] = pd.to_datetime(
    df["Time"],
    format="%H:%M:%S",
    errors="coerce"
).dt.hour
```

```python
df.head()
```

Create Time Slot

```python
def get_time_slot(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df["time_slot"] = df["booking_hour"].apply(get_time_slot)
```
